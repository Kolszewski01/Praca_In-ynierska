1. **MLPRegressor (Multi-layer Perceptron Regressor):**
   - Jest to model regresji oparty na sztucznych sieciach neuronowych, składający się z co najmniej trzech warstw: warstwy wejściowej, ukrytej (lub warstw) oraz warstwy wyjściowej. 
   - Model ten może być stosowany do problemów regresji, gdzie dąży się do przewidywania ciągłych wartości wyjściowych.

2. **RandomForestRegressor:**
   - Jest to model regresji oparty na algorytmie lasów losowych, który składa się z wielu drzew decyzyjnych. 
   - Każde drzewo w lesie jest trenowane na różnych podzbiorach danych, co prowadzi do zrównoważonej i silnej generalizacji. 
   - Model ten jest elastyczny i nadaje się do wielu rodzajów danych, przy zachowaniu odporności na overfitting.

3. **DecisionTreeRegressor:**
   - Jest to model regresji oparty na drzewach decyzyjnych, który dzieli zbiór danych na podzbiory, stosując serię reguł decyzyjnych opartych na wartościach cech. 
   - Model ten buduje drzewo decyzyjne, które jest używane do prognozowania wartości ciągłych na podstawie danych wejściowych. 
   - Może być podatny na overfitting, ale może być regulowany przez parametryzację.

4. **LinearRegression:**
   - Jest to podstawowy model regresji liniowej, który zakłada liniową zależność między zmiennymi niezależnymi a zmienną zależną. 
   - Model ten stara się dopasować linię najlepszego dopasowania do danych poprzez minimalizację sumy kwadratów różnic między wartościami przewidywanymi a rzeczywistymi. 
   - Jest łatwy w interpretacji i implementacji, ale może być ograniczony, gdy dane nie spełniają założeń liniowości.


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.svm import SVR
```

```python
# REGRESJA LINIOWA

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('Błąd średniokwadratowy:', mse)

print('Współczynniki regresji:', model.coef_)
print('Wyraz wolny:', model.intercept_)
print('Współczynnik determinacji (R-squared):', r2)
```

```python
# DRZEWA DECYZYJNE

model = DecisionTreeRegressor()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print('Błąd średniokwadratowy:', mse)

r2 = r2_score(y_test, y_pred)
print('Współczynnik determinacji (R-squared):', r2)

 
mae = mean_absolute_error(y_test, y_pred)
print('Średni błąd bezwzględny (MAE):', mae)

rmse = np.sqrt(mse)
print('Średni błąd kwadratowy pierwiastkowany (RMSE):', rmse)

```

```python
# LAS LOSOWY

model = RandomForestRegressor()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print('Błąd średniokwadratowy:', mse)

r2 = r2_score(y_test, y_pred)
print('Współczynnik determinacji (R-squared):', r2)

 
mae = mean_absolute_error(y_test, y_pred)
print('Średni błąd bezwzględny (MAE):', mae)

rmse = np.sqrt(mse)
print('Średni błąd kwadratowy pierwiastkowany (RMSE):', rmse)

```

```python
# SIECI NEURONOWE

model = MLPRegressor(hidden_layer_sizes=(100, 50,20), max_iter=500, alpha=0.001, early_stopping=True)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print('Błąd średniokwadratowy:', mse)

r2 = r2_score(y_test, y_pred)
print('Współczynnik determinacji (R-squared):', r2)

 
mae = mean_absolute_error(y_test, y_pred)
print('Średni błąd bezwzględny (MAE):', mae)

rmse = np.sqrt(mse)
print('Średni błąd kwadratowy pierwiastkowany (RMSE):', rmse)

print('Strata po ostatniej epoce:', model.loss_)

```

1. **Błąd średniokwadratowy (MSE):** Średni kwadrat różnicy między przewidywanymi a rzeczywistymi wartościami.

2. **Współczynnik determinacji (R-squared):** Miara, która określa, jak dobrze model pasuje do danych, porównując wariancję przewidywań modelu do wariancji rzeczywistych danych.

3. **Średni błąd bezwzględny (MAE):** Średnia wartość bezwzględna różnicy między przewidywanymi a rzeczywistymi wartościami.

4. **Średni błąd kwadratowy pierwiastkowany (RMSE):** Pierwiastek kwadratowy z MSE, co daje miarę przeciętnego błędu przewidywania w tych samych jednostkach, co oryginalne dane.

5. **Strata po ostatniej epoce:** Wartość błędu oszacowana przez model podczas ostatniej epoki treningowej.




## IMPLEMENTACJA MODELU

### 1. Przesyłanie danych do API
System pobiera dane dotyczące aktualnej temperatury i wilgotności poprzez api NASA oraz api GIOS

### 2. Przetwarzanie przez model
API Django odbiera dane, przetwarza je (np. konwertuje na odpowiedni format) i używa wcześniej zapisanego modelu do wykonania predykcji. Model ocenia, czy dane warunki mogą prowadzić do pożaru.

### 3. Zwracanie wyniku
API zwraca wynik predykcji do interfejsu użytkownika React, który może to być na przykład wartość binarna (0 lub 1, gdzie 1 oznacza w ryzyko pożaru).

### 4. Wyświetlanie wyniku użytkownikowi
Aplikacja React odbiera odpowiedź z API i wyświetla użytkownikowi wynik, informując, czy istnieje ryzyko pożaru.

```python
# IMPORT MODELU

import joblib

joblib.dump(model, 'model.pkl')


# POBIERANIE DANYCH - PRZYKŁAD 

#PLIK DATA_FETCHER

import requests

def get_nasa_fires_data():
    url = "https://nasa.gov/fires_api/latest_fires"
    response = requests.get(url)
    return response.json() if response.status_code == 200 else None

def get_weather_data(location):
    api_key = "weatherapi_key"
    url = f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={location}"
    response = requests.get(url)
    return response.json() if response.status_code == 200 else None

def get_gios_air_quality_data(station_id):
    url = f"http://api.gios.gov.pl/pjp-api/rest/aqindex/getIndex/{station_id}"
    response = requests.get(url)
    return response.json() if response.status_code == 200 else None


# ZAŁADOWANIE MODELU | DEFINICJA FUNKCJI

model = joblib.load('path/model.pkl')

def predict_fire_risk(temperature, humidity):
    input_data = np.array([[temperature, humidity]])
    prediction = model.predict(input_data)
    return prediction[0]


# UZYCIE FUNKCJI W WIDOKU API DJANGO

from django.http import JsonResponse
from rest_framework.views import APIView
from .data_fetcher import get_nasa_fires_data, get_weather_data, get_gios_air_quality_data
from .model import predict_fire_risk, prepare_data

class FireRiskPredictionAPIView(APIView):
    def get(self, request, *args, **kwargs):
        fire_data = get_nasa_fires_data()
        weather_data = get_weather_data('Location')

        if fire_data and weather_data
            input_data = prepare_data(fire_data, weather_data)
            prediction = predict_fire_risk(input_data)
            return JsonResponse({'prediction': prediction})
        else:
            return JsonResponse({'error': 'Error fetching data'}, status=400)

        
```
